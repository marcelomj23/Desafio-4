```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_ipeadata = pd.read_csv('../data/ipeadata_last.csv', sep=';', encoding='latin-1')
df_ipeadata.head()
```

```python
# rename columns to Preço
df_ipeadata.rename(columns={'Preço - petróleo bruto - Brent (FOB)': 'Preco'}, inplace=True)
```

```python
# fix dtypes
df_ipeadata['Data'] = pd.to_datetime(df_ipeadata['Data'], format='%Y-%m-%d')
df_ipeadata['Preco'] = df_ipeadata['Preco'].astype(float)
df_ipeadata.dtypes
```

```python
# describe
df_ipeadata.describe()
```

```python
# plot
df_ipeadata.plot(x='Data', y='Preco', figsize=(15, 5), title='Preço do petróleo bruto - Brent (FOB)')
plt.ylim(0, None)
```

```python
# plot only year 2023 y-axis star at 0 step 10
df_ipeadata[df_ipeadata['Data'].dt.year == 2023].plot(x='Data', y='Preco', figsize=(15, 5), title='Preço do petróleo bruto - Brent (FOB)')
plt.ylim(0, None)
plt.yticks(range(0, 100, 10))
plt.show()
```
